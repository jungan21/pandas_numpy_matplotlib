### 数据分组 Groupby

```python
# header=0 表示第0行是column 名字
# nrows = 1000 表示只读取1000行，如果没有nrows, 就读取全部行
df = pd.read_csv('./data/ad_feature.csv', sep=',', header=0, nrows=1000)
cate_group = df.groupby('cate_id')    # 按照cate_id进行分组，把cate_id相同的数据分成一组
# 也可以df.groupby(['cate_id', 'campaign_id'])
type(cate_group)
```

```python
# print每一类 有多少个
cate_group.size()
```

```python
cate_group['price', 'brand'].mean().head()    # 计算每一组price和brand的平均值，必须保证函数在每个组中产生一个值
```

```python
cate_group['price', 'brand'].mean().reset_index(drop=False).head()    # reset_index会重新设置index，drop参数控制是否丢弃原来的index
```

### 异常值 / 离群值（Outlier Data）处理

```python
price_des = df['price'].describe()
price_des
```

两种方法：
1. 可以用mean 和 std 来计算，根据经验吗98%的数据落在 (min, max )
max = mean + 3 * std
min = mean - 3 * std
2. 也可以用分位数来算，例如下面的数据处理

```python
# 这里用分位数来clip
# max: 50% + 3* ( 75%- 50%)
# min: 50% - 3* ( 50%- 25%)
valid_max = price_des['50%'] + 3 * (price_des['75%'] - price_des['50%'])
valid_min = price_des['50%'] - 3 * (price_des['50%'] - price_des['25%'])
# clip 把大于最大值的都用最大值替换，把小于最小值的都用最小值替换
df['price'] = df['price'].clip(valid_min, valid_max)    # 把异常值变为边界值
df['price'].describe()
```
